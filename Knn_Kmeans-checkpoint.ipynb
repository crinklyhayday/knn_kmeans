```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cluster import KMeans

# Load training and testing datasets
train_df = pd.read_csv(r"C:\Users\Shreyas\Downloads\traink.csv")
test_df = pd.read_csv(r"C:\Users\Shreyas\Downloads\testk.csv")

# Split training dataset into features (X) and target (y)
X_train = train_df.iloc[:, :-1]
y_train = train_df.iloc[:, -1]

# Separate numeric and non-numeric columns
num_cols = X_train.select_dtypes(include=[np.number]).columns
cat_cols = X_train.select_dtypes(exclude=[np.number]).columns

# Create an instance of the SimpleImputer class to handle missing values
imputer = SimpleImputer(strategy='mean')
X_train[num_cols] = imputer.fit_transform(X_train[num_cols])
X_test = test_df.copy()
X_test[num_cols] = imputer.transform(X_test[num_cols])

# Convert all column names to strings
X_train.columns = X_train.columns.astype(str)
X_test.columns = X_test.columns.astype(str)

# Create a separate OneHotEncoder for each categorical variable
encoders = {}
for col in cat_cols:
    # Get unique categories for the current column
    unique_vals = X_train[col].unique().tolist()
    # Convert all categories to strings
    unique_vals = [str(val) for val in unique_vals]
    # Sort the categories
    unique_vals = sorted(unique_vals)
    # Create a new OneHotEncoder for the current column
    encoders[col] = OneHotEncoder(categories=[unique_vals], handle_unknown='ignore')
    # Fit and transform the current column
    encoded_col = encoders[col].fit_transform(X_train[[col]])
    # Replace the original column with the one-hot encoded column
    X_train = pd.concat([X_train.drop(col, axis=1), pd.DataFrame(encoded_col.toarray())], axis=1)

# Convert all column names to strings
X_train.columns = X_train.columns.astype(str)

# Train KNN model
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

# Preprocess test data
for col in cat_cols:
    # Fit and transform the current column
    encoded_col = encoders[col].transform(X_test[[col]])
    # Replace the original column with the one-hot encoded column
    X_test = pd.concat([X_test.drop(col, axis=1), pd.DataFrame(encoded_col.toarray())], axis=1)

# Convert all column names to strings
X_test.columns = X_test.columns.astype(str)

# Use trained model to make predictions on the test data
predictions = knn.predict(X_test)
```
