```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cluster import KMeans

train_df = pd.read_csv(r"C:\Users\Shreyas\Downloads\traink.csv")        #enter training dataset file path downloaded from kaggle
test_df = pd.read_csv(r"C:\Users\Shreyas\Downloads\testk.csv")          #enter testing dataset file path downloaded from kaggle 

X_train = train_df.iloc[:, :-1]
y_train = train_df.iloc[:, -1]

num_cols = X_train.select_dtypes(include=[np.number]).columns
cat_cols = X_train.select_dtypes(exclude=[np.number]).columns

imputer = SimpleImputer(strategy='mean')
X_train[num_cols] = imputer.fit_transform(X_train[num_cols])
X_test = test_df.copy()
X_test[num_cols] = imputer.transform(X_test[num_cols])

X_train.columns = X_train.columns.astype(str)
X_test.columns = X_test.columns.astype(str)

encoders = {}
for col in cat_cols:
    unique_vals = X_train[col].unique().tolist()
    unique_vals = [str(val) for val in unique_vals]
    unique_vals = sorted(unique_vals)
    encoders[col] = OneHotEncoder(categories=[unique_vals], handle_unknown='ignore')
    encoded_col = encoders[col].fit_transform(X_train[[col]])
    X_train = pd.concat([X_train.drop(col, axis=1), pd.DataFrame(encoded_col.toarray())], axis=1)

X_train.columns = X_train.columns.astype(str)

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

for col in cat_cols:
    encoded_col = encoders[col].transform(X_test[[col]])
    X_test = pd.concat([X_test.drop(col, axis=1), pd.DataFrame(encoded_col.toarray())], axis=1)

X_test.columns = X_test.columns.astype(str)

predictions = knn.predict(X_test)
```

```python
X_train_split, X_val, y_train_split, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)
knn.fit(X_train_split, y_train_split)
val_preds = knn.predict(X_val)
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_val, val_preds)
print(f"Accuracy: {accuracy*100}%")
from math import sqrt
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_val, val_preds)
rmse = sqrt(mse)

print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
```

```python
from sklearn.ensemble import RandomForestRegressor
num_clusters = 3
rf = RandomForestRegressor(n_jobs=-1)
rf.fit(X_train, y_train)
importances = rf.feature_importances_
top_n_features = np.argsort(importances)[::-1][:5]
selected_features = X_train.iloc[:, top_n_features]

kmeans = KMeans(n_clusters=num_clusters)
kmeans.fit(selected_features)
train_clusters = kmeans.predict(selected_features)
test_clusters = kmeans.predict(X_test.iloc[:, top_n_features])

train_df['cluster'] = train_clusters
test_df['cluster'] = test_clusters

import matplotlib.pyplot as plt
plt.figure(figsize=(10, 8))
colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k']
for i in range(num_clusters):
    indices = np.where(train_clusters == i)[0]
    plt.scatter(selected_features.iloc[indices, 1], selected_features.iloc[indices, 2], c=colors[i % len(colors)], label=f'Cluster {i+1}')

plt.title('Clusters of training data')
plt.xlabel(X_train.columns[top_n_features[1]])
plt.ylabel(X_train.columns[top_n_features[2]])
plt.legend()
plt.show()
```
